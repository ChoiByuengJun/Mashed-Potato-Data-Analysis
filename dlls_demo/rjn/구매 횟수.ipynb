```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

#   
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux  
    plt.rc('font', family='NanumGothic')

#    
plt.rcParams['axes.unicode_minus'] = False

# CSV  
file_path = 'preprocessed_data.csv'  # CSV   
df = pd.read_csv(file_path)

#    
if ' ' in df.columns:
    # :     
    plt.figure(figsize=(10, 6))
    plt.hist(df[' '], bins=20, edgecolor='black')
    plt.title('  ')
    plt.xlabel(' ')
    plt.ylabel('')
    plt.grid(True)
    plt.show()

    # :    
    if '' in df.columns:
        avg_purchase_by_category = df.groupby('')[' '].mean().sort_values()
        plt.figure(figsize=(10, 6))
        avg_purchase_by_category.plot(kind='bar', color='skyblue')
        plt.title('   ')
        plt.xlabel('')
        plt.ylabel('  ')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()

    # :     
    if ' ' in df.columns:
        avg_purchase_by_size = df.groupby(' ')[' '].mean().sort_values()
        plt.figure(figsize=(10, 6))
        avg_purchase_by_size.plot(kind='bar', color='lightcoral')
        plt.title('    ')
        plt.xlabel(' ')
        plt.ylabel('  ')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()

else:
    print(" ' '  .")

```
