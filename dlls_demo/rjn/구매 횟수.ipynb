```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '구매 횟수' in df.columns:
    # 히스토그램: 고객별 총 구매 횟수 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['구매 횟수'], bins=20, edgecolor='black')
    plt.title('구매 횟수 분포')
    plt.xlabel('구매 횟수')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

    # 막대그래프: 업종별 평균 구매 횟수
    if '업종' in df.columns:
        avg_purchase_by_category = df.groupby('업종')['구매 횟수'].mean().sort_values()
        plt.figure(figsize=(10, 6))
        avg_purchase_by_category.plot(kind='bar', color='skyblue')
        plt.title('업종별 평균 구매 횟수')
        plt.xlabel('업종')
        plt.ylabel('평균 구매 횟수')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()

    # 막대그래프: 회사 규모별 평균 구매 횟수
    if '회사 규모' in df.columns:
        avg_purchase_by_size = df.groupby('회사 규모')['구매 횟수'].mean().sort_values()
        plt.figure(figsize=(10, 6))
        avg_purchase_by_size.plot(kind='bar', color='lightcoral')
        plt.title('회사 규모별 평균 구매 횟수')
        plt.xlabel('회사 규모')
        plt.ylabel('평균 구매 횟수')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()

else:
    print("데이터프레임에 '구매 횟수' 열이 없습니다.")

```

0회에서 5회 사이의 구매 횟수를 가진 고객이 가장 많습니다.
구매 횟수가 높아질수록 고객 수는 급격히 감소하는 추세를 보입니다.
최대 30회 정도의 구매 횟수가 존재하지만, 이러한 경우는 드뭅니다.
대다수의 고객은 소량 구매를 선호합니다. 이는 구매 주기, 상품 특성, 서비스 이용 패턴 등 다양한 요인과 관련될 수 있습니다.
일부 고객은 반복적으로 많은 횟수의 구매를 진행하며, 이는 충성 고객일 가능성이 큽니다.

대부분의 업종에서 평균 구매 횟수가 비슷한 수준으로 8 ~ 9회로 유지됩니다.
업종에 따른 큰 차이는 관찰되지 않지만, 특정 업종은 약간 더 높은 경향을 보입니다.
고객은 업종에 관계없이 일정한 빈도로 구매를 진행하는 경향이 있습니다.
특정 업종에 대한 더 많은 구매가 나타나지 않는다는 점에서, 업종별 차별화된 혜택이나 프로모션 전략이 필요할 수 있습니다.

소기업은 평균 구매 횟수가 2회 미만으로 가장 낮은 수준입니다.
중소기업과 중견기업의 평균 구매 횟수는 약 5회 ~ 6회로 비슷한 수준입니다.
대기업의 평균 구매 횟수는 12회 이상으로 가장 높습니다.
대기업 고객은 중소기업 및 소기업 고객에 비해 훨씬 더 빈번하게 구매를 진행합니다.
대규모 구매를 선호하는 고객층에 대한 추가적인 혜택 제공이나 유지 전략이 필요할 수 있습니다.

대다수의 고객은 적은 횟수의 구매를 진행하지만, 특정 규모(대기업)에서는 높은 구매 빈도를 보입니다.
업종별로는 큰 차이가 없지만, 회사 규모에 따라 평균 구매 횟수의 큰 차이가 존재합니다.
대기업 고객군을 대상으로 하는 맞춤형 관리와 소규모 고객군의 이탈 방지 대책이 중요합니다.
