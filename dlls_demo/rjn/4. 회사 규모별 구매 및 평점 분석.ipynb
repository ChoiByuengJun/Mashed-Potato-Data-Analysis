```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

#  
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

#  
file_path = 'preprocessed_data.csv'
data = pd.read_csv(file_path)

# 4.      
avg_purchase_by_company_size = data.groupby(' ')['  '].mean()
plt.figure(figsize=(12, 6))
avg_purchase_by_company_size.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('     ', fontsize=16)
plt.xlabel(' ', fontsize=12)
plt.ylabel('    ()', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=' ', y=' ', data=data, palette='Set3', hue=' ', dodge=False)
plt.title('    ', fontsize=16)
plt.xlabel(' ', fontsize=12)
plt.ylabel(' ', fontsize=12)
plt.tight_layout()
plt.show()
```
