```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '평균 평점' in df.columns and '이탈 여부' in df.columns:
    # 히스토그램: 평균 평점 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['평균 평점'], bins=10, edgecolor='black', color='skyblue')
    plt.title('평균 평점 분포')
    plt.xlabel('평균 평점')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

    # 박스플롯: 이탈 여부에 따른 평균 평점 비교 (palette 제거)
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='이탈 여부', y='평균 평점', data=df)
    plt.title('이탈 여부에 따른 평균 평점 비교')
    plt.xlabel('이탈 여부 (0: 유지, 1: 이탈)')
    plt.ylabel('평균 평점')
    plt.grid(True)
    plt.show()

else:
    print("데이터프레임에 '평균 평점' 또는 '이탈 여부' 열이 없습니다.")
```

관찰된 패턴:
평균 평점 3점 ~ 3.5점 구간에 가장 많은 고객이 분포해 있습니다.
**극단적인 평점 (1점 또는 5점)**을 기록한 고객은 상대적으로 적습니다.
정규분포 형태에 가까운 분포를 보이며, 중간 평점에 몰리는 경향이 강합니다.
결론:
대부분의 고객은 **중간 수준(3점 ~ 3.5점)**의 평가를 주며, 서비스에 대해 보통 수준의 만족도를 보이는 경우가 많습니다.
극단적으로 높은 점수(5점) 또는 낮은 점수(1점)를 주는 고객은 비교적 적습니다.

관찰된 패턴:
유지 고객(0): 평균 평점의 중앙값이 약 3점이며, 점수 분포가 비교적 좁은 범위에서 나타납니다.
이탈 고객(1): 평균 평점의 중앙값이 약 3.5점으로 유지 고객보다 약간 높지만, 점수 분포가 넓으며 1점 또는 5점 같은 극단적인 점수가 많이 존재합니다.
이상치: 유지 고객과 이탈 고객 모두 1점 및 5점의 이상치가 존재합니다.
결론:
유지 고객은 **안정적으로 중간 수준(3점)**의 평가를 주는 경우가 많습니다.
이탈 고객은 평균 평점이 다양한 범위에 걸쳐 있으며, **긍정적 평가(4점 이상)**를 주고도 이탈하는 경우가 나타납니다. 이는 고객 만족도가 높았음에도 불구하고 외부 요인(경쟁 서비스, 가격 등)으로 이탈했을 가능성이 있습니다.
이탈 고객 중 **부정적 평가(1점 ~ 2점)**를 남긴 경우는 서비스 불만족으로 인한 이탈일 수 있습니다.

종합 결론:
평균 평점이 높다고 해서 반드시 유지되는 것은 아니며, 이탈 고객 중 일부는 높은 점수를 남기고도 이탈합니다.
서비스 개선과 충성 고객 관리뿐만 아니라, 경쟁 서비스 대응 및 가격 정책이 중요할 수 있습니다.
평균 평점이 낮은 고객을 사전 식별하여 이탈 방지를 위한 맞춤형 전략이 필요합니다.
