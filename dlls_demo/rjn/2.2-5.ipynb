```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
data = pd.read_csv('preprocessed_data.csv')

# 지역별 구매 횟수 합계 계산
region_purchases = data.groupby('지역')['구매 횟수'].sum()

# 원그래프로 시각화
plt.figure(figsize=(8, 8))
region_purchases.plot(kind='pie', autopct='%1.1f%%')
plt.title('지역별 구매 횟수 비율')
plt.ylabel('')  # y축 레이블 제거
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['평균 구매 금액'], data['평균 평점'], alpha=0.2, s=10)  # 투명도 및 점 크기 조정
plt.xscale('log')  # x축을 로그 스케일로 설정
plt.xlabel('평균 구매 금액 (KRW) - 로그 스케일')
plt.ylabel('평균 평점')
plt.title('평균 구매 금액과 평균 평점의 상관관계')
plt.grid(True)
plt.show()
```

데이터 집중: 대부분의 데이터 포인트가 10^8 KRW 이하의 구매 금액에 집중되어 있으며, 평균 평점은 대체로 3.0 이상입니다.
상관관계 부재: 높은 구매 금액과 높은 평점 간에 명확한 상관관계가 보이지 않습니다.
이상치 존재: 매우 높은 구매 금액을 보이는 소수의 데이터 포인트가 존재하며, 이는 분석에서 이상치로 고려될 수 있습니다.
이러한 관찰을 통해 평균 구매 금액이 평균 평점에 직접적인 영향을 미치지 않을 수 있음을 추론할 수 있으며, 이는 구매 금액이 높다고 해서 반드시 고객 만족도가 높은 것은 아니라는 점을 시사합니다.

```python
# 가입 기간에 따른 총 구매 금액 분포 시각화
plt.figure(figsize=(10, 6))
plt.hist(data['총 구매 금액'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('총 구매 금액 (KRW)')
plt.ylabel('빈도')
plt.title('가입 기간에 따른 총 구매 금액 분포')
plt.grid(True)
plt.show()
```

```python

# 그래프 크기 설정
plt.figure(figsize=(16, 8))  # 그래프의 크기를 더 크게 설정

# 상자그래프 시각화
data.boxplot(column='평균 평점', by='구매 주기', showfliers=False, widths=0.6)  # 각 박스의 너비 조정
plt.xlabel('구매 주기')
plt.ylabel('평균 평점')
plt.title('구매 주기 대비 평균 평점')
plt.suptitle('')  # 기본 제공되는 서브타이틀을 제거
plt.xticks(rotation=45)  # x축 레이블을 45도로 회전하여 가독성 향상
plt.grid(True)  # 그리드 추가
plt.show()
```

평균 평점의 중앙값: 대부분의 구매 주기에서 중앙값은 약 3.5로, 고객들이 대체로 중간 이상의 만족도를 보고하고 있음을 나타냅니다.
이상치의 존재: 많은 구매 주기 범주에서 특히 낮은 평점의 이상치가 눈에 띕니다, 이는 일부 고객들이 특정 구매 주기에서 매우 낮은 만족도를 경험하고 있음을 반영할 수 있습니다.
평점 범위: 대부분의 평점은 3.0에서 4.0 사이에 위치하지만, 구매 주기에 따라 평점이 2.5 이하로 떨어지는 경우도 있어, 구매 주기에 따라 고객 만족도의 변동성이 큼을 보여줍니다.
데이터 밀도와 변동성: 상자그래프의 높이가 변동적이며, 이는 구매 주기별 고객 활동 및 평점의 밀도와 변동성을 나타냅니다.
