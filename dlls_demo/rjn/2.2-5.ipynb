```python
import pandas as pd
import matplotlib.pyplot as plt

#  
data = pd.read_csv('preprocessed_data.csv')

#     
region_purchases = data.groupby('')[' '].sum()

#  
plt.figure(figsize=(8, 8))
region_purchases.plot(kind='pie', autopct='%1.1f%%')
plt.title('   ')
plt.ylabel('')  # y  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['  '], data[' '], alpha=0.2, s=10)  #     
plt.xscale('log')  # x   
plt.xlabel('   (KRW) -  ')
plt.ylabel(' ')
plt.title('     ')
plt.grid(True)
plt.show()
```

```python
#        
plt.figure(figsize=(10, 6))
plt.hist(data['  '], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('   (KRW)')
plt.ylabel('')
plt.title('      ')
plt.grid(True)
plt.show()
```
