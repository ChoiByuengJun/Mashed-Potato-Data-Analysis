```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '총 구매 금액' in df.columns:
    # 히스토그램: 총 구매 금액 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['총 구매 금액'], bins=20, edgecolor='black', color='skyblue')
    plt.title('총 구매 금액 분포')
    plt.xlabel('총 구매 금액')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

    # 산점도: 총 구매 금액과 평균 평점의 관계
    if '평균 평점' in df.columns:
        plt.figure(figsize=(10, 6))
        plt.scatter(df['총 구매 금액'], df['평균 평점'], alpha=0.5, color='coral')
        plt.title('총 구매 금액과 평균 평점의 관계')
        plt.xlabel('총 구매 금액')
        plt.ylabel('평균 평점')
        plt.grid(True)
        plt.show()

    # 산점도: 총 구매 금액과 업종 간의 관계
    if '업종' in df.columns:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x='총 구매 금액', y='업종', data=df, alpha=0.7)
        plt.title('총 구매 금액과 업종 간의 관계')
        plt.xlabel('총 구매 금액')
        plt.ylabel('업종')
        plt.grid(True)
        plt.show()

else:
    print("데이터프레임에 '총 구매 금액' 열이 없습니다.")
```

0원에서 1억 원대의 총 구매 금액을 가진 고객이 가장 많습니다.
2억 원 이상의 구매 금액을 가진 고객은 급격히 감소하며, 5억 원 이상의 총 구매 금액을 가진 고객은 드문 경우입니다.
대다수의 고객은 상대적으로 낮은 총 구매 금액을 기록하고 있습니다.
일부 고객만 고액을 지출하며, 이들은 주요 VIP 또는 대량 구매 고객일 가능성이 큽니다.

총 구매 금액이 낮은 경우에도 평균 평점 4~5점을 주는 경우가 많습니다.
고액을 지출한 고객들 중에도 평균 평점이 1점인 경우가 일부 존재합니다.
평균 평점이 3점 ~ 4점에 집중되어 있으며, 총 구매 금액이 높아질수록 평점 분포는 여전히 균일합니다.
총 구매 금액과 평균 평점 사이에 강한 상관관계는 관찰되지 않습니다.
높은 금액을 지출한다고 해서 반드시 높은 평점을 주는 것은 아니며, 고객 만족도는 구매 금액 이외의 다른 요인에 의해 결정되는 경우가 많습니다.

대부분의 업종에서 총 구매 금액이 0원에서 1억 원 사이에 집중됩니다.
일부 업종(예: 건설/산업 장비 대여 업체, 물류/운송회사)에서는 4억~6억 원대의 대량 구매가 존재합니다.
반면에, 공유 차량 서비스나 택시회사 등 서비스 중심 업종은 고액의 총 구매 금액을 기록하는 경우가 거의 없습니다.
업종에 따라 총 구매 금액의 분포가 다르며, 제품이나 서비스의 특성에 따라 고액 구매 여부가 달라집니다.
대여 및 물류 업종은 대량 거래가 흔한 반면, 서비스형 업종은 상대적으로 낮은 구매 금액 범위를 유지합니다.

대다수의 고객은 비교적 낮은 금액(1억 원 이하)의 구매를 선호합니다.
총 구매 금액이 높아질수록 고객의 평점은 여전히 다양하며, 고객 만족도는 총 지출 금액보다는 서비스 품질과 같은 다른 요인에 더 큰 영향을 받을 가능성이 큽니다.
대형 거래를 유도할 수 있는 업종별 전략과 고액 고객 관리 방안이 중요할 수 있습니다.
