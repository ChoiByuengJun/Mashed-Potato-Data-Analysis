```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
data = pd.read_csv('preprocessed_data.csv')

# 업종별 평균 구매 금액 계산
industry_means = data.groupby('업종')['평균 구매 금액'].mean()

# 그래프 크기 설정
plt.figure(figsize=(12, 6))  # 가로 12인치, 세로 6인치 크기의 그래프 생성

# 막대그래프로 시각화
industry_means.plot(kind='bar', color='skyblue')
plt.xlabel('업종')  # x축 레이블
plt.ylabel('평균 구매 금액 (KRW)')  # y축 레이블
plt.title('업종별 평균 구매 금액')  # 그래프 제목
plt.xticks(rotation=30)  # x축 레이블을 30도로 회전
plt.show()
```

업종별로 구매 금액의 평균을 시각화 (막대그래프)
