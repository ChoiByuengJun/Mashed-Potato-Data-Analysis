```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

#  
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

#  
file_path = 'preprocessed_data.csv'
data = pd.read_csv(file_path)

# 5.      
region_counts = data[''].value_counts()
total_purchase_by_region = data.groupby('')['  '].sum()

plt.figure(figsize=(12, 6))
region_counts.head(10).plot(kind='bar', color='purple', edgecolor='black')
plt.title(' 10   ', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel(' ', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
total_purchase_by_region.sort_values(ascending=False).head(10).plot(kind='bar', color='brown', edgecolor='black')
plt.title(' 10    ', fontsize=16)
plt.xlabel('', fontsize=12)
plt.ylabel('   ()', fontsize=12)
plt.tight_layout()
plt.show()
```
