```python
import pandas as pd
import matplotlib.pyplot as plt
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '구매 가격' in df.columns and '평균 평점' in df.columns:
    
    # 히스토그램: 구매 가격 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['구매 가격'], bins=30, edgecolor='black')
    plt.title('구매 가격 분포')
    plt.xlabel('구매 가격')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

    # 산점도: 구매 가격과 평균 평점의 관계
    plt.figure(figsize=(10, 6))
    plt.scatter(df['구매 가격'], df['평균 평점'], alpha=0.5)
    plt.title('구매 가격과 평균 평점의 관계')
    plt.xlabel('구매 가격')
    plt.ylabel('평균 평점')
    plt.grid(True)
    plt.show()

else:
    print("데이터프레임에 '구매 가격' 또는 '평균 평점' 열이 없습니다.")


```

일반적인 거래 금액은 1억 ~ 2억 원대가 주를 이루며, 고가의 거래는 드물게 발생합니다.
구매 가격이 고르게 분포되지 않고 특정 구간에 집중된 경향이 있습니다.

평균 평점과 구매 가격 사이에 강한 선형 관계는 관찰되지 않습니다.
특정 구매 금액대에서 평점이 균일하지 않고, 저평가 및 고평가가 함께 나타나므로, 구매 금액 자체가 고객 만족도를 결정하는 주 요인이 아닐 가능성이 있습니다.
평균적으로 대부분의 구매는 중간 수준의 평점(3~4점)을 유지하고 있습니다.

구매 금액이 클수록 고객 만족도가 반드시 높아지거나 낮아지는 경향은 보이지 않습니다.
거래량은 1억 ~ 2억 원대가 가장 많으며, 이 가격대가 주된 구매 금액 범위입니다.
평점이 높은 경우와 낮은 경우가 모두 존재하므로, **다른 요인(예: 서비스 품질, 제품 문제 등)**이 고객 평점에 영향을 주고 있을 수 있습니다.
