```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

# 폰트 설정
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터 로드
file_path = 'preprocessed_data.csv'
data = pd.read_csv(file_path)

# 6. 가입 일자별 고객 및 분포 분석
data['가입 연도'] = pd.to_datetime(data['가입 일자']).dt.year
yearly_customer_counts = data['가입 연도'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
yearly_customer_counts.plot(kind='line', marker='o', color='blue')
plt.title('가입 연도별 고객 수 변화', fontsize=16)
plt.xlabel('가입 연도', fontsize=12)
plt.ylabel('고객 수', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.hist(data['가입 연도'], bins=10, color='grey', edgecolor='black')
plt.title('가입 연도 분포', fontsize=16)
plt.xlabel('가입 연도', fontsize=12)
plt.ylabel('빈도', fontsize=12)
plt.tight_layout()
plt.show()
```
