```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '구매 주기' in df.columns and '총 구매 금액' in df.columns:
    # 히스토그램: 구매 주기 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['구매 주기'], bins=20, edgecolor='black', color='skyblue')
    plt.title('구매 주기 분포')
    plt.xlabel('구매 주기 (일)')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

    # 산점도: 구매 주기와 총 구매 금액의 관계
    plt.figure(figsize=(10, 6))
    plt.scatter(df['구매 주기'], df['총 구매 금액'], alpha=0.5, color='coral')
    plt.title('구매 주기와 총 구매 금액의 관계')
    plt.xlabel('구매 주기 (일)')
    plt.ylabel('총 구매 금액')
    plt.grid(True)
    plt.show()

else:
    print("데이터프레임에 '구매 주기' 또는 '총 구매 금액' 열이 없습니다.")
```

1. 히스토그램 (구매 주기 분포)
해석:
X축(구매 주기): 고객의 평균 구매 주기(일 단위).
Y축(빈도수): 해당 구매 주기를 가진 고객 수.
관찰된 패턴:
0 ~ 50일 사이: 대다수의 고객이 짧은 구매 주기(50일 이내)를 유지합니다.
200일 이상의 구매 주기를 가진 고객은 매우 드뭅니다.
일부 고객은 1000일 이상의 매우 긴 구매 주기를 가지고 있지만, 극히 소수에 해당합니다.
결론:
대부분의 고객은 비교적 짧은 기간 내에 반복적으로 구매합니다.
긴 구매 주기를 가진 고객은 서비스를 비정기적으로 이용하거나 특정 이벤트에 따라 구매하는 경우일 수 있습니다.
고객의 구매 빈도 유지를 위한 프로모션이나 리마인드 알림 전략이 필요할 수 있습니다.
2. 산점도 (구매 주기와 총 구매 금액의 관계)
해석:
X축(구매 주기): 고객의 평균 구매 주기.
Y축(총 구매 금액): 해당 고객의 총 구매 금액.
관찰된 패턴:
**짧은 구매 주기(0 ~ 50일)**를 가진 고객 중 일부는 매우 높은 총 구매 금액을 기록합니다.
구매 주기가 길어질수록 총 구매 금액이 0원에 수렴하는 경향이 강합니다.
200일 이상의 긴 구매 주기를 가진 고객의 대부분은 총 구매 금액이 낮은 편입니다.
결론:
짧은 구매 주기를 유지하는 고객일수록 총 구매 금액이 높을 가능성이 큽니다.
긴 구매 주기를 가진 고객은 구매 횟수가 적어 총 구매 금액이 낮아지는 경향이 있습니다.
반복 구매를 유도할 수 있는 혜택 제공이나 마케팅 캠페인을 통해 고객의 구매 주기를 단축할 필요가 있습니다.
종합 결론:
대부분의 고객은 짧은 구매 주기를 보이며, 이들은 주로 높은 총 구매 금액을 기록합니다.
긴 구매 주기를 가진 고객은 비정기적인 구매 패턴을 보이며, 총 구매 금액이 낮습니다.
구매 주기가 긴 고객을 대상으로 리마인드 메시지를 보내거나 주기적인 프로모션을 통해 반복 구매를 유도하면 구매 주기 단축 및 총 구매 금액 증가를 기대할 수 있습니다.
