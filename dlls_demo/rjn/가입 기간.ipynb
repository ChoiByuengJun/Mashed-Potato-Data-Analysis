```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '가입 기간' in df.columns and '평균 구매 금액' in df.columns:
    # 산점도: 가입 기간과 평균 구매 금액
    plt.figure(figsize=(10, 6))
    plt.scatter(df['가입 기간'], df['평균 구매 금액'], alpha=0.5, color='coral')
    plt.title('가입 기간과 평균 구매 금액의 관계')
    plt.xlabel('가입 기간 (일)')
    plt.ylabel('평균 구매 금액')
    plt.grid(True)
    plt.show()

    # 히스토그램: 가입 기간 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['가입 기간'], bins=20, edgecolor='black', color='skyblue')
    plt.title('가입 기간 분포')
    plt.xlabel('가입 기간 (일)')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

else:
    print("데이터프레임에 '가입 기간' 또는 '평균 구매 금액' 열이 없습니다.")
```

관찰된 패턴:
가입 기간 0 ~ 500일 사이: 많은 고객이 분포되어 있으며, 평균 구매 금액은 0원에서 최대 6억 원까지 넓은 범위로 분포됩니다.
가입 기간 500일 이후: 고객 수는 점점 줄어들지만, 여전히 다양한 금액대의 구매가 이루어지고 있습니다.
고가 구매: 가입 기간이 길수록 고가의 평균 구매 금액이 소수 존재하지만, 전체적으로는 특정한 패턴을 보이지 않습니다.
결론:
가입 기간이 길어진다고 해서 평균 구매 금액이 반드시 증가하거나 감소하지는 않습니다.
고객의 평균 구매 금액은 가입 기간보다는 다른 요인(업종, 이벤트 등)에 의해 결정될 가능성이 큽니다.
충성 고객 관리를 통해 장기 고객의 고가 구매를 유도할 필요가 있습니다.
