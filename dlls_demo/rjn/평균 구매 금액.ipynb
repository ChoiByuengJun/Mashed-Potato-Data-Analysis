```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform

# 한글 폰트 설정
if platform.system() == 'Darwin':  # Mac OS
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # Windows
    plt.rc('font', family='Malgun Gothic')
else:  # Linux 또는 기타
    plt.rc('font', family='NanumGothic')

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# CSV 파일 불러오기
file_path = 'preprocessed_data.csv'  # CSV 파일 경로 입력
df = pd.read_csv(file_path)

# 데이터 확인 및 시각화
if '평균 구매 금액' in df.columns:
    # 히스토그램: 평균 구매 금액 분포
    plt.figure(figsize=(10, 6))
    plt.hist(df['평균 구매 금액'], bins=20, edgecolor='black', color='skyblue')
    plt.title('평균 구매 금액 분포')
    plt.xlabel('평균 구매 금액')
    plt.ylabel('빈도수')
    plt.grid(True)
    plt.show()

    # 박스플롯: 업종별 평균 구매 금액 (palette 제거)
    if '업종' in df.columns:
        plt.figure(figsize=(12, 8))
        sns.boxplot(x='업종', y='평균 구매 금액', data=df)
        plt.title('업종별 평균 구매 금액')
        plt.xlabel('업종')
        plt.ylabel('평균 구매 금액')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()

    # 박스플롯: 회사 규모별 평균 구매 금액
    if '회사 규모' in df.columns:
        plt.figure(figsize=(12, 8))
        sns.boxplot(x='회사 규모', y='평균 구매 금액', data=df)
        plt.title('회사 규모별 평균 구매 금액')
        plt.xlabel('회사 규모')
        plt.ylabel('평균 구매 금액')
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()

else:
    print("데이터프레임에 '평균 구매 금액' 열이 없습니다.")

```

1. 히스토그램 (평균 구매 금액 분포)
해석:
X축(평균 구매 금액): 고객별 평균 거래당 구매 금액.
Y축(빈도수): 해당 평균 구매 금액을 가진 고객 수.
관찰된 패턴:
0원에서 1억 원대의 평균 구매 금액을 가진 고객이 대부분을 차지합니다.
평균 구매 금액이 2억 원 이상인 고객은 상대적으로 적습니다.
소수의 고객은 4억 원 이상의 고액 평균 구매 금액을 기록하고 있습니다.
결론:
고객의 대다수는 1억 원 이하의 평균 구매 금액을 기록하며, 일부 고액 거래 고객이 존재합니다.
VIP 고객 또는 고액 구매 고객을 대상으로 하는 맞춤형 혜택 및 충성도 프로그램이 효과적일 수 있습니다.
2. 박스플롯 (업종별 평균 구매 금액)
해석:
X축(업종): 각 업종 이름.
Y축(평균 구매 금액): 해당 업종의 평균 구매 금액.
관찰된 패턴:
물류/운송업과 건설 장비 대여 업종은 평균 구매 금액의 중앙값이 높고, 고액 구매 사례도 많이 포함되어 있습니다.
운전학원, 렌터카 업체 등 서비스형 업종의 평균 구매 금액은 상대적으로 낮은 수준입니다.
각 업종별로 이상치(고액 구매)가 많이 존재하며, 특히 중고차 딜러 업종에서도 고액 거래가 발생합니다.
결론:
특정 업종(건설 및 물류 관련)은 대형 거래가 많으며, 서비스형 업종은 상대적으로 소액 거래가 많습니다.
서비스형 업종의 구매 금액을 높이기 위해 패키지 상품이나 장기 계약 프로모션을 도입할 필요가 있습니다.
3. 박스플롯 (회사 규모별 평균 구매 금액)
해석:
X축(회사 규모): 소기업, 중소기업, 대기업 등 회사 규모.
Y축(평균 구매 금액): 해당 회사 규모의 평균 구매 금액.
관찰된 패턴:
대기업 고객의 평균 구매 금액 중앙값이 가장 높습니다.
소기업의 평균 구매 금액은 중앙값이 낮고, 이상치가 많지 않습니다.
중소기업과 중견기업은 평균 구매 금액이 대체로 비슷하며, 1억 원에서 2억 원대에 해당합니다.
결론:
대기업 고객은 상대적으로 높은 평균 구매 금액을 보이지만, 소규모 고객은 낮은 평균 구매 금액을 보입니다.
소규모 고객군의 구매 금액을 증대시키기 위해 장기 계약 할인 또는 다회 구매 혜택을 제공하는 것이 필요할 수 있습니다.
종합 결론:
대다수 고객은 1억 원 이하의 평균 구매 금액을 기록하지만, 일부 업종(물류, 건설 등)에서는 고액 거래가 빈번하게 이루어집니다.
대기업 고객은 소기업 고객에 비해 높은 구매 금액을 보이며, VIP 관리 전략이 중요합니다.
서비스형 업종과 소기업 고객을 대상으로 프로모션 및 맞춤형 전략을 통해 구매 금액을 증대할 기회가 존재합니다.
